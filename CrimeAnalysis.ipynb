```python
import os
import pandas as pd
import folium
os.getcwd()
#os.chdir('/Users/dnchawla/Downloads')
path = '2018.csv'
df = pd.read_csv(path)

df.head()

new_df = df.loc[:,['Year', 'District', 'Primary Type', 'Arrest', 'Latitude', 'Longitude']]

df2023 = new_df[new_df['Year'] == 2023]
#makes 2023 dataframe
len(df2023.index)
#So far in 2023, there has been 61,254 crimes committed

df2022 = new_df[new_df['Year'] == 2022]
#makes 2021 dataframe
len(df2022.index)
#In 2022, there has been 237,880 crimes committed

df2021 = new_df[new_df['Year'] == 2021]
#makes 2021 dataframe
len(df2021.index)
#In 2021, there has been 208,539 crimes committed

df2020 = new_df[new_df['Year'] == 2020]
#makes 2020 dataframe
len(df2020.index)
#In 2020, there has been 212,076 crimes committed

df2019 = new_df[new_df['Year'] == 2019]
#makes 2020 dataframe
len(df2019.index)
#In 2019, there has been 261,237 crimes committed

df2018 = new_df[new_df['Year'] == 2018]
#makes 2018 dataframe
len(df2018.index)
#In 2018, there has been 268,767 crimes committed

#Line chart showing number of crimes per year
import matplotlib.pyplot as plt
year = ['2018', '2019', '2020', '2021', '2022']
crime = [len(df2018), len(df2019), len(df2020), len(df2021), len(df2022)]
plt.plot(year, crime)
plt.title("Crimes per Year (2018-2022)")
plt.xlabel("Year")
plt.ylabel("Crimes per Year")
plt.show()

# 2018-2023 Responses for Primary Type
value_counts = new_df['Primary Type'].value_counts()
crimes = pd.DataFrame(value_counts)
print(value_counts)
type(value_counts)

plt.plot(crimes)
plt

len(df)

print(df.columns)

crimes.plot.pie(y='Primary Type', figsize=(8, 8), autopct='%1.1f%%', startangle=90)
plt.legend().remove()
plt.title('Crime Types')
plt.show()

districts = df.loc[:,['District', 'Arrest']]

district_count = districts.groupby('District')['Arrest'].sum()


district_count

district_count.plot.pie(x = 'Districts', y='Arrest', figsize=(100, 10), autopct='%1.1f%%', startangle=90)
plt.title('Arrests per Districts (2018-2023)')

arrests_per_type = df.loc[:,['Primary Type', 'Arrest']]
arresttype = arrests_per_type.groupby('Primary Type')['Arrest'].sum()
type(arresttype)

arresttype




```

```python
df19 = df[df['Year'] == 2019]

# Drop any rows with null values in the Latitude and Longitude columns
df19.dropna(subset=['Latitude', 'Longitude'], inplace=True)

# Find the center of the map
location_center = [df19['Latitude'].mean(), df19['Longitude'].mean()]

# Create a map
m = folium.Map(location=location_center, zoom_start=11)

# Add markers for each crime incident location
for index, row in df19.iterrows():
    folium.Marker(location=[row['Latitude'], row['Longitude']]).add_to(m)

# Display the map
m.save('map.html')
```
